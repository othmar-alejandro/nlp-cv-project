This notebook demonstrates the **ReAct (Reasoning + Acting)** framework using LangChain agents.  
The agent combines a search tool and a math tool to answer multi-step questions.

## Step 2: Import Libraries & Load API Keys

````python
import openai
import os

# Updated imports for LangChain 1.x package split
from langchain_openai import OpenAI
from langchain_community.agent_toolkits.load_tools import load_tools
from langchain_classic.agents import initialize_agent
from dotenv import load_dotenv

# Load environment variables from .env file
load_dotenv()

# Load API keys
os.environ["OPENAI_API_KEY"] = os.getenv("OPENAI_API_KEY")
os.environ["SERPER_API_KEY"] = os.getenv("SERPER_API_KEY")

print("✅ Libraries imported and API keys loaded.")
````

## Step 3: Configure the LLM, Tools, and Agent

We configure:
- **LLM**: `gpt-3.5-turbo-instruct` (replacement for deprecated `text-davinci-003`)
- **Tools**: `google-serper` for web search + `llm-math` for calculations
- **Agent**: `zero-shot-react-description` — the ReAct framework agent

````python
# Initialize the LLM
# Note: text-davinci-003 is deprecated; gpt-3.5-turbo-instruct is the current equivalent
llm = OpenAI(model_name="gpt-3.5-turbo-instruct", temperature=0)

# Load tools: web search + math calculator
tools = load_tools(["google-serper", "llm-math"], llm=llm)

# Initialize the ReAct agent
agent = initialize_agent(
    tools,
    llm,
    agent="zero-shot-react-description",
    verbose=True
)

print("✅ Agent initialized successfully.")
````

## Step 4: Run the Agent

The agent will:
1. **Search** for who Olivia Wilde's boyfriend is
2. **Search** for his age
3. **Calculate** his age raised to the 0.23 power

No few-shot examples needed — the ReAct framework handles all reasoning autonomously.

````python
agent.run("Who is Olivia Wilde's boyfriend? What is his current age raised to the 0.23 power?")
````
